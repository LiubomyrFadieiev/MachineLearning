```python
import numpy as np
import timeit as t 
import random

def Multy(A,B):
    n = len(A)
    AB = np.zeros((n,n))
    for i in range(n):
        for k in range(n):
            for j in range(n):
                AB[i][j] += A[i][k] * B[k][j]
    return AB

def MultyNP(A,B):
    n = len(A)
    AB = np.zeros((n,n))
    AB = A.dot(B)
    return AB

def Random(n):
    A = np.zeros((n,n))
    min = 0
    max = 10
    for i in range(n):
        for j in range(n):
            sign = np.sign(random.randint(min,max)-3)
            number = random.randint(min,max)
            A[i][j] = sign*number
    return A
```
